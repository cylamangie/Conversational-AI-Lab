```python
# ================== Exercise 3: RAG example ==================
import ollama
client = ollama.Client()
from llama_index.core import Document, VectorStoreIndex
from llama_index.embeddings.ollama import OllamaEmbedding
from llama_index.core.settings import Settings

### Copied from RAG example ###

data = [
    """Module code: COM6104
Module name: Topics in Data Science and AI
Lecturer: Dr. WK Wong""",
    """Module code: COM1001
Module name: Introduction to AI
Lecturer: Dr. AB Chan""",
    """Module code: COM1002
Module name: Advanced AI
Lecturer: Dr. CD Cheung""",
    """Module code: COM1003
Module name: Expert AI
Lecturer: Dr. EF Chen""",
    """Module code: COM1004
Module name: Profession AI
Lecturer: Dr. GH Lam""",
]
documents = [Document(text=text) for text in data]

Settings.embed_model = OllamaEmbedding("qwen3-embedding")

index = VectorStoreIndex.from_documents(documents)
retriever = index.as_retriever(similarity_top_k=2)

### --- End --- ###


# To convert the conversation into a text display in the prompt
def get_msg(msg: dict):
    if msg["role"] == "assistant":
        return f"AI: {msg["content"]}"
    return f"User: {msg["content"]}"


def update_sys_prompt(history: list):
    # With query rewriting
    msg_list = [get_msg(msg) for msg in history if msg["role"] != "system"]
    # The conversation will keep the latest conversation between the AI and user
    conversation = "\n".join(msg_list)
    rewrite_prompt = f"""Read the following conversation.
Rewrtie the last user message for a RAG query.
Just output the rewritten message without any format.

Conversation
===
{conversation}
==="""
    # Ask the AI to rewrite the message to make its context clear in the retrieval
    response = client.chat(
        model="qwen3",
        messages=[{"role": "user", "content": rewrite_prompt}],
        think=False,
        options={"temperature": 0},
    )
    rewritten_query = response.message.content

    # Printing the rewritten query for reference
    # This is not necessary in the application
    print("RAG query:", rewritten_query)

    retrieved_nodes = retriever.retrieve(rewritten_query)
    kb = [f"{pos+1}. {item.node.text}" for pos, item in enumerate(retrieved_nodes)]
    "\n\n".join(kb)

    return f"""You are an admin of Hang Seng University.
Refer the following knowledge to answer students' enquiry about the university.
Keep your answer short and to the point.

Knowledge
===
{kb}
==="""


sys_msg = {"role": "system", "content": ""}

done = False
history = [sys_msg]
while not done:
    msg = input()
    done = msg == "exit"
    if not done:
        # Use the chat history to ask the AI to rewrite the user message!
        history.append({"role": "user", "content": msg})
        sys_msg["content"] = update_sys_prompt(history)
        print("User:", msg)

        response = client.chat(
            model="qwen3",
            messages=history.copy(),
            think=False,
            options={"temperature": 0},
        )

        history.append(response.message)
        history = [history[0], *history[1:][-20:]]

        print("AI:", response.message.content)
```
